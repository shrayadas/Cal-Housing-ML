```python
import numpy as np

indices = np.random.choice(california_housing.frame.index, size=500, replace=False)
```

```python
sns.scatterplot(
    data=california_housing.frame.loc[indices],
    x="Longitude",
    y="Latitude",
    hue="MedHouseVal",
    palette="plasma",  #optional
    alpha=0.5
)
plt.title("Median house value by location")
plt.tight_layout()
```

```python
import pandas as pd
subset = california_housing.frame.loc[indices].drop(columns=["Longitude", "Latitude"])
```

```python
subset["MedHouseVal"] = pd.qcut(subset["MedHouseVal"], 6).map(lambda x: x.mid)
```

```python
sns.pairplot(data=subset, hue="MedHouseVal", palette="viridis")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate

alphas = np.logspace(-3, 1, 30)
model = make_pipeline(StandardScaler(), RidgeCV(alphas=alphas))

cv_results = cross_validate(
    model,
    california_housing.data,
    california_housing.target,
    return_estimator=True,
    n_jobs=2,
)
```

```python
score = cv_results["test_score"]
print(f"R2 score: {score.mean():.3f}  {score.std():.3f}")
```

```python
import pandas as pd
coefs = pd.DataFrame(
    [est.named_steps["ridgecv"].coef_ for est in cv_results["estimator"]],
    columns=california_housing.feature_names
)
```

```python
coefs.plot.box(vert=False)
plt.axvline(0, linestyle="--")
plt.title("Ridge model coefficients")
```
